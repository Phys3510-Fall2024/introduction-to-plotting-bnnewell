```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider # import the interact function and the sliders we need for this environemtn
```

```python
def plot_contour( k1=1, k2=1, c=0, tatertot=0, levels= 10):
    #defining the maxs and mins 
    x_min = -6 
    y_min = -6
    x_max =  6
    y_max =  6
    
    #generating the grid
    q1 = np.linspace(x_min, x_max, 400)
    q2 = np.linspace(y_min, y_max, 400)
    Q1 , Q2 = np.meshgrid(q1,q2)

    
    #defining function
    Z = 0.5*k1 * Q1 ** 2 + 0.5 * k2 * Q2 ** 2 + c * Q1 * Q2 ** 2
    
    #time for plotting
    plt.figure(figsize = ( 5 , 5 ))
    contour_filled = plt.contourf(Q1, Q2, Z, levels = levels, cmap = 'plasma')
    contour_lines  = plt.contour(Q1, Q2, Z, levels = levels, colors = 'green', linewidths = 0.5)
    #adding lables
    plt.clabel(contour_lines, inline=True, fontsize= 10)
    #axis lines
    plt.xlabel('$q_1$', fontsize = 12)
    plt.ylabel('$q_2$', fontsize = 12)
    plt.xlim([x_min,x_max])
    plt.ylim([y_min,y_max])
    #we has arrows for some reason
    arrow_length = 0.3* (y_max - y_min)
    plt.arrow(0, 0, arrow_length * np.cos(2*np.pi*tatertot/360), arrow_length * np.sin(2*np.pi*tatertot/360),
              color='red', head_width=0.2)
    plt.arrow(0, 0, -arrow_length * np.sin(2*np.pi*tatertot/360), arrow_length * np.cos(2*np.pi*tatertot/360),
              color='red', head_width=0.2)
    #show the plot
    plt.show()
```

```python
plot_contour(1, 1, 1, 15, levels=10)
```

```python
 

interact(plot_contour,
         k1=FloatSlider(min=-2, max=2, step=0.1, value=1, description='k1'),
         k2=FloatSlider(min=-2, max=2, step=0.1, value=1, description='k2'),
         c=FloatSlider(min=-2, max=2, step=0.1, value=0, description='c'),
         tatertot=FloatSlider(min=0, max=360, step=1, value=0, description='tatertot'),
         levels=IntSlider(min=5, max=20, step=1, value=10, description='Levels')       
        )

plt.show()
```
