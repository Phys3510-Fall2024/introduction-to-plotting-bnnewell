```python
# Importing matplotlib and numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace( 0,10,5)
print(x)
```

```python
def parabola(x): #defining function
    return x**2
#valuses will be equal
y_parabola = parabola(x)

print(y_parabola)
```

```python
# print([x,y_parabola])
print(np.transpose([x,y_parabola]))
```

```python
plt.plot(x, y_parabola) #plotting the line)

plt.show() #gets rid of the info on top
```

```python
x = np.linspace(-10, 10, 50) # 50 points between -10 and 10
y_parabola = parabola(x) # update y_parabola
```

```python
plt.plot(x,y_parabola)
plt.show()
```

```python
plt.plot(x,y_parabola)
plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.show()
```

```python
plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5') #playing with the look


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.show()
```

```python
plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5', marker='o',
         markersize=5,markerfacecolor='#06d6fd') #playing with the look


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.show()
```

```python
plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5', marker='o',
         markersize=5,markerfacecolor='#06d6fd') #playing with the look


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.grid(True, color='gray', linestyle='--', linewidth=0.5) #(adding the grid
plt.show()
```

```python
plt.figure(figsize=(5, 3)) #changing the size of the figure

plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5', marker='o',
         markersize=5,markerfacecolor='#06d6fd') #playing with the look


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.grid(True, color='gray', linestyle='--', linewidth=0.5) #(adding the grid


plt.xlim(-4,4)
plt.ylim(-2,10)

#font tick lables
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)

plt.show()
```

```python
#line defining
def linear(x):
    return -3 * x
#line y values

y_linear = linear(x)


plt.figure(figsize=(5, 3)) #changing the size of the figure

plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5', marker='o',
         markersize=5,markerfacecolor='#06d6fd') #playing with the look
plt.plot(x,y_linear, color='blue')


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.grid(True, color='gray', linestyle='--', linewidth=0.5) #(adding the grid


plt.xlim(-4,4)
plt.ylim(-2,10)

#font tick lables
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)

plt.show()
```

```python
#line defining
def linear(x):
    return -3 * x
#line y values

y_linear = linear(x)


plt.figure(figsize=(8, 3)) #changing the size of the figure

plt.subplot(1,2,1)
plt.plot(x,y_parabola, color='green',linestyle = '--', linewidth  = '5', marker='o',
         markersize=5,markerfacecolor='#06d6fd') #playing with the look

plt.xlabel("testing units but on a different graph") #title for x derection of graph
plt.ylabel("testing but sideways on a different graph") #units for y
plt.grid(True, color='gray', linestyle='--', linewidth=0.5) #(adding the grid

plt.subplot(1,2,2)
plt.plot(x,y_linear, color='blue')


plt.xlabel("testing units going this derection") #title for x derection of graph
plt.ylabel("testing but sideways") #units for y
plt.grid(True, color='gray', linestyle='--', linewidth=0.5) #(adding the grid


plt.xlim(-4,4)
plt.ylim(-2,10)

#font tick lables
plt.xticks(fontsize=15) 
plt.yticks(fontsize=15)

plt.tight_layout()

plt.show()
```
